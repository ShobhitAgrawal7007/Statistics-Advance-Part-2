1. What is hypothesis testing in statistics?

Hypothesis testing in statistics is a method used to assess a claim about a population parameter using sample data. It involves formulating a null hypothesis (a statement of no effect or no difference) and an alternative hypothesis (a statement of what the researcher believes to be true). By analyzing sample data, we determine whether there's enough evidence to reject the null hypothesis in favor of the alternative.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

The null hypothesis is a statement of "no effect" or "no difference," while the alternative hypothesis is a statement that contradicts the null hypothesis, suggesting there is a significant effect or difference.

Key differences are:

1. Null Hypothesis (H0):

(A). Definition: The null hypothesis is a statement that assumes there is no relationship, no effect, or no difference between variables in a population.

(B). Purpose: It serves as the default position that researchers try to disprove.

(C). Example: "There is no difference in test scores between students who study with flashcards and those who don't."

2. Alternative Hypothesis (Ha or H1):

(A). Definition: The alternative hypothesis is a statement that contradicts the null hypothesis and suggests there is a relationship, effect, or difference.

(B). Purpose: It's the claim that researchers are trying to find evidence for.

(C). Example: "Students who study with flashcards will have significantly higher test scores than those who don't."

3. What is the significance level in hypothesis testing, and why is it important?

In hypothesis testing, the significance level (alpha, α) is the probability of rejecting the null hypothesis when it is actually true. It represents the researcher's predetermined threshold for how much risk they're willing to take in incorrectly rejecting the null hypothesis (a false positive). A lower significance level indicates a stricter standard for rejecting the null hypothesis, requiring stronger evidence to do so.

Why is Significance Level Important?
1. Controlling Type I Error:
The significance level directly controls the risk of making a Type I error (false positive). By setting α, researchers determine how likely they are to mistakenly conclude there's a significant effect when there isn't.
2. Balancing Risk:
Choosing the significance level involves a trade-off. A lower α reduces the risk of Type I errors but increases the risk of failing to reject a false null hypothesis (Type II error). A higher α increases the risk of Type I errors but decreases the risk of Type II errors.
3. Contextual Decision:
The appropriate significance level depends on the context of the research. In fields like medical research, where the consequences of a false positive can be severe, a lower α (e.g., 0.01) might be chosen. In exploratory studies, a higher α (e.g., 0.10) might be more appropriate to avoid missing potential effects.
4. Setting the Standard:
The significance level establishes the standard for statistical significance. It helps researchers determine whether their findings are likely due to a real effect or just random chance.

4. What does a P-value represent in hypothesis testing?

In hypothesis testing, a p-value represents the probability of obtaining results as extreme as, or more extreme than, the results observed in a study, assuming the null hypothesis is true.

Here's a more detailed explanation:
1. Null Hypothesis:
In hypothesis testing, the null hypothesis is a statement that there is no effect or relationship between the variables being studied.
2. Observed Results:
These are the results obtained from a sample in a study.
3. P-value Calculation:
The p-value is calculated based on the observed results and the assumption that the null hypothesis is true.
4. Interpretation:
A small p-value (typically below a predetermined threshold like 0.05) suggests that the observed results are unlikely to have occurred by chance alone, if the null hypothesis were true. This leads to the rejection of the null hypothesis in favor of an alternative hypothesis. Conversely, a large p-value suggests that the observed results are likely due to chance, and the null hypothesis cannot be rejected.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

1. One-Tailed Test:

(A). Directional Hypothesis: The alternative hypothesis specifies a direction (e.g., "greater than" or "less than").

(B). Critical Region: The critical region (where the test statistic must fall to reject the null hypothesis) is located entirely on one side of the distribution.

(C). Example: Testing if the average height of men is greater than 5'10".

2. Two-Tailed Test:

(A). Non-Directional Hypothesis: The alternative hypothesis simply states that the parameter is different from a specific value (e.g., "not equal to").

(B). Critical Region: The critical region is split between both sides of the distribution.

(C). Example: Testing if the average height of men is different from 5'10" (could be taller or shorter).

Key Differences:

1. Direction:
One-tailed tests are directional, while two-tailed tests are not.
2. Critical Region:
One-tailed tests have the critical region on one side, two-tailed tests have it on both.
3. Power:
One-tailed tests are more powerful for detecting a difference in the specific direction predicted, but less powerful for detecting a difference in the opposite direction.
4. Sample Size:
Two-tailed tests generally require a larger sample size to achieve the same level of statistical power as a one-tailed test, according to Statsig.

Choosing the Right Test:

1. One-tailed test: Use when you have a strong prior belief or hypothesis about the direction of the effect.
2. Two-tailed test: Use when you are interested in any difference, regardless of direction.

9. How do you calculate the Z-score, and what does it represent in hypothesis testing?

A z-score measures how many standard deviations a data point is away from the mean of a dataset. In hypothesis testing, it represents the test statistic, indicating how far the sample mean deviates from the hypothesized population mean, in terms of standard errors.

Calculating the Z-score:

The formula for calculating a z-score is:

z = (x - μ) / σ

Where:

z: is the z-score.

x: is the data point (or sample mean in hypothesis testing).

μ: is the population mean.

σ: is the population standard deviation.

If the population standard deviation is unknown, it can be estimated using the sample standard deviation (s), and the formula becomes: z = (x - μ) / (s / √n), where n is the sample size.

Z-score in Hypothesis Testing:

In hypothesis testing, the z-score helps determine if the observed difference between a sample and a population (or between two samples) is statistically significant. Here's how:
1. Calculate the z-score:
You calculate the z-score using the sample mean, population mean, and standard deviation (or sample standard deviation and sample size). This z-score represents how many standard errors the sample mean is away from the hypothesized population mean.
2. Determine the critical value:
Based on the chosen significance level (alpha) and whether it's a one-tailed or two-tailed test, you find the critical z-value(s) from a z-table.
3. Compare:
If the calculated z-score falls outside the range of the critical values (i.e., it's more extreme), you reject the null hypothesis. This indicates that the observed difference is unlikely to have occurred by chance alone, and there is evidence to support the alternative hypothesis. If the calculated z-score falls within the range of the critical values, you fail to reject the null hypothesis.

10. What is the T-distribution, and when should it be used instead of the normal distribution?

The t-distribution, also known as Student's t-distribution, is a probability distribution used in statistics, particularly when dealing with small sample sizes or when the population standard deviation is unknown. It is a suitable alternative to the normal distribution in these scenarios because it accounts for the uncertainty introduced by estimating the population standard deviation from a sample. The t-distribution is bell-shaped and symmetric like the normal distribution, but it has heavier tails, meaning it has a higher probability of extreme values.

Why Use the T-Distribution Instead of the Normal Distribution in These Cases?
1. Normal Distribution Assumption:
The normal distribution assumes you know the population standard deviation, which is often unrealistic in practical scenarios, especially with small samples.
2. Overestimation of Precision:
When using the normal distribution with estimated standard deviations from small samples, you might overestimate the precision of your results, leading to potentially misleading conclusions.
3. T-distribution's Heavier Tails:
The t-distribution's heavier tails account for the increased uncertainty associated with estimating the population standard deviation from a small sample, providing a more accurate representation of the data's variability.
4. More Conservative Results:
The t-distribution generally leads to wider confidence intervals and more conservative hypothesis test results compared to using the normal distribution with estimated standard deviations, which can be more realistic when dealing with uncertainty.

12. What is the T-test, and how is it used in hypothesis testing?

A t-test is a statistical hypothesis test used to determine if there's a significant difference between the means of two groups.

Here's how it's used in hypothesis testing:
1. Define Hypotheses:
A t-test begins with formulating null and alternative hypotheses. The null hypothesis (H0) typically assumes no significant difference between the means of the groups being compared. The alternative hypothesis (Ha) suggests there is a significant difference.
2. Choose a t-test type:
There are different types of t-tests, including independent samples t-tests (for comparing two independent groups) and paired samples t-tests (for comparing related groups, like before-and-after measurements on the same individuals).
3. Calculate the t-statistic:
The t-test involves calculating a t-statistic based on the sample data. This statistic quantifies the difference between the means relative to the variability within the groups.
4. Determine the p-value:
The calculated t-statistic is then used to find a p-value. The p-value represents the probability of observing a difference as extreme as, or more extreme than, the one calculated, assuming the null hypothesis is true.
5. Make a decision:
If the p-value is below a predetermined significance level (alpha, typically 0.05), the null hypothesis is rejected, suggesting a statistically significant difference between the means. If the p-value is greater than alpha, the null hypothesis is not rejected.
6. Interpret results:
The results of a t-test help determine if a treatment, intervention, or any other factor has a meaningful impact on the groups being compared.

14. What is a confidence interval, and how is it used to interpret statistical results?

A confidence interval (CI) is a range of values that is likely to contain the true value of an unknown population parameter, like a mean or proportion. It's used to quantify the uncertainty associated with an estimate derived from a sample of data. The confidence level associated with the interval (e.g., 95%, 99%) indicates the probability that the interval will contain the true population parameter if the sampling process were repeated many times.

How it's used to interpret statistical results:
1. Understanding the Precision of an Estimate:
A confidence interval provides a range of plausible values for the population parameter, rather than just a single point estimate. This range gives an indication of the precision of the estimate. A narrower interval suggests a more precise estimate, while a wider interval indicates more uncertainty.
2. Assessing the Significance of Results:
If a confidence interval for the difference between two groups (e.g., the effect of a treatment) does not include zero, it suggests that the observed difference is unlikely to be due to random chance and is statistically significant.
3. Making Inferences about the Population:
By providing a range of plausible values, confidence intervals allow researchers to make inferences about the population based on sample data. For example, a 95% confidence interval for the average height of students in a class can be used to estimate the average height of all students in the school.
4. Comparing Results Across Studies:
Confidence intervals can be used to compare results from different studies. If the confidence intervals for the same parameter in two studies overlap, it suggests that the results are consistent. If they do not overlap, it suggests a statistically significant difference between the studies.
5. Evaluating the Reliability of Predictions:
In areas like finance, confidence intervals can be used to assess the reliability of predictions. For example, a confidence interval for future stock prices can help investors understand the potential range of outcomes.

18. What is the Chi-square goodness of fit test, and how is it applied?

The chi-square goodness-of-fit test is a statistical hypothesis test used to determine how well a sample distribution matches an expected distribution. It assesses whether there's a significant difference between observed frequencies (what you actually see in your data) and expected frequencies (what you would anticipate based on a specific distribution).

Here's how it's applied:

1. Define the Hypothesis:

(A). Null Hypothesis (H0): The observed distribution matches the expected distribution.

(B). Alternative Hypothesis (Ha): The observed distribution does not match the expected distribution.
2. Collect and Organize Data: Gather your observed data, grouping it into categories or intervals.
3. Calculate Expected Frequencies: For each category, determine the expected frequency based on the hypothesized distribution.
4. Calculate the Chi-Square Statistic: Use the formula: χ² = Σ [(Observed - Expected)² / Expected]. This formula calculates the sum of squared differences between observed and expected frequencies, divided by the expected frequencies.
5. Determine Degrees of Freedom: Degrees of freedom (df) are calculated as (number of categories - 1).
6. Find the P-value: Using the calculated chi-square statistic and degrees of freedom, find the corresponding p-value from a chi-square distribution table or statistical software.
7. Make a Decision:

(A). If the p-value is less than your chosen significance level (alpha, often 0.05), reject the null hypothesis. This suggests a statistically significant difference between the observed and expected distributions.

(B). If the p-value is greater than or equal to the significance level, fail to reject the null hypothesis. This indicates that the observed data is consistent with the expected distribution.

19. What is the F-distribution, and when is it used in hypothesis testing?

The F-distribution is a probability distribution that is used in hypothesis testing, particularly when comparing variances or means of multiple groups. It's commonly used in analysis of variance (ANOVA) and in testing the significance of regression models. The F-distribution helps determine if observed differences between groups are statistically significant or likely due to random chance.

Uses of the F-distribution in Hypothesis Testing:
1. Analysis of Variance (ANOVA):
ANOVA uses the F-distribution to test the null hypothesis that the means of two or more groups are equal. It compares the variance between groups to the variance within groups. If the F-statistic (calculated from the variances) is large enough, the null hypothesis is rejected, suggesting that at least one group mean is significantly different from the others.
2. Comparing Variances:
The F-distribution is used to test if two sample variances are significantly different from each other. This is done by calculating an F-statistic which is the ratio of the two sample variances.
3. Regression Analysis:
In regression, the F-test is used to assess the overall significance of the model. It compares the variance explained by the model to the unexplained variance (error). A significant F-statistic suggests that the model as a whole is a good fit for the data.

20. What is an ANOVA test, and what are its assumptions?

An ANOVA (Analysis of Variance) test is a statistical method used to compare the means of three or more groups to determine if there is a statistically significant difference between them.

Assumptions of ANOVA:

To ensure the validity of ANOVA results, several key assumptions need to be met:
1. Normality:
The data within each group should be approximately normally distributed. This means that the data should follow a bell-shaped curve when plotted.
2. Homogeneity of Variance:
The variances (or standard deviations) of the groups being compared should be roughly equal. This is also known as homoscedasticity.
3. Independence:
The observations within each group and between groups should be independent of each other. This means that the data collected from one group should not influence the data collected from another group.

The primary types of ANOVA tests are one-way ANOVA and two-way ANOVA. One-way ANOVA is used when there's one independent variable with multiple levels (groups) influencing a dependent variable. Two-way ANOVA is used when there are two independent variables, allowing analysis of their individual effects and their interaction on the dependent variable.

Here's a breakdown of the two main types:
1. One-Way ANOVA:

Purpose: Compares the means of three or more groups based on a single independent variable.

Example: Comparing the average test scores of students in three different teaching methods.

Key feature: Only one independent variable is considered.
2. Two-Way ANOVA:

Purpose:
Examines the effects of two independent variables and their interaction on a dependent variable.

Example:
Investigating the effect of both teaching method (one independent variable) and student gender (another independent variable) on test scores.

Key feature:
Allows analysis of individual (main) effects and the interaction effect between the two independent variables.

2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define parameters for the discrete uniform distribution
low = 1  # inclusive lower bound
high = 7  # exclusive upper bound (like a die roll: 1 to 6)

# Generate discrete uniform random variables
data = randint.rvs(low, high, size=1000)

# Define possible values
x = np.arange(low, high)

# Compute PMF
pmf = randint.pmf(x, low, high)

# Plot PMF
plt.stem(x, pmf)
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('PMF of Discrete Uniform Distribution (1 to 6)')
plt.grid(True)
plt.show()
```

3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution?

```python
def bernoulli_pdf(x, p):
    """
    Calculate the PDF (PMF) of a Bernoulli distribution.

    Parameters:
        x (int): The outcome (0 or 1).
        p (float): Probability of success (0 ≤ p ≤ 1).

    Returns:
        float: Probability of observing value x.
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for a Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1.")

    return p**x * (1 - p)**(1 - x)

# Example usage
print("P(X=0):", bernoulli_pdf(0, 0.3))
print("P(X=1):", bernoulli_pdf(1, 0.3))
```

4. Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram?

5. Create a Poisson distribution and visualize it using Python?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters
lambda_ = 4  # average rate (λ)
size = 1000  # number of samples

# Generate Poisson-distributed data
data = np.random.poisson(lam=lambda_, size=size)

# Define range for plotting
x = np.arange(0, max(data)+1)

# Compute theoretical PMF
pmf = poisson.pmf(x, mu=lambda_)

# Plot histogram of simulated data
plt.hist(data, bins=np.arange(0, max(data)+2)-0.5, density=True, alpha=0.6, edgecolor='black', label='Simulated')

# Plot theoretical PMF
plt.plot(x, pmf, 'ro-', label='Theoretical PMF')

# Formatting
plt.title('Poisson Distribution (λ = 4)')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.xticks(x)
plt.legend()
plt.grid(True)
plt.show()
```

6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the discrete uniform distribution parameters
low = 1   # inclusive lower bound
high = 7  # exclusive upper bound (for values 1 through 6)

# Define values where to compute the CDF
x = np.arange(low, high + 1)

# Calculate the CDF using scipy
cdf = randint.cdf(x, low, high)

# Plot the CDF
plt.step(x, cdf, where='post', label='CDF', color='blue', linewidth=2)
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.title('CDF of Discrete Uniform Distribution (1 to 6)')
plt.xticks(x)
plt.ylim(0, 1.1)
plt.grid(True)
plt.legend()
plt.show()
```

15. Implement hypothesis testing using Z-statistics for a sample dataset?

```python
import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    n = len(sample_data)
    sample_mean = np.mean(sample_data)

    # Step 1: Calculate the Z-statistic
    z = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Step 2: Compute the p-value for two-tailed test
    p_value = 2 * (1 - norm.cdf(abs(z)))

    # Step 3: Compare with significance level (alpha)
    print("=== Hypothesis Test Using Z-Statistic ===")
    print(f"Sample Mean        = {sample_mean:.3f}")
    print(f"Population Mean    = {population_mean}")
    print(f"Z-statistic        = {z:.3f}")
    print(f"P-value            = {p_value:.4f}")

    if p_value < alpha:
        print(f"Result: Reject the null hypothesis (significant at α = {alpha})")
    else:
        print(f"Result: Fail to reject the null hypothesis (not significant at α = {alpha})")

# Example usage:
np.random.seed(42)
sample = np.random.normal(loc=51, scale=10, size=50)  # Sample with mean close to 51
population_mu = 50
population_sigma = 10

z_test(sample, population_mu, population_sigma)
```

16. Create a confidence interval for a dataset using Python and interpret the result?

```python
import numpy as np
from scipy.stats import norm

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # Sample standard deviation

    # Z-critical value for desired confidence level
    z_critical = norm.ppf(1 - (1 - confidence) / 2)

    # Margin of error
    margin_of_error = z_critical * (std / np.sqrt(n))

    # Confidence interval
    lower = mean - margin_of_error
    upper = mean + margin_of_error

    print("=== Confidence Interval Calculation ===")
    print(f"Sample Size           = {n}")
    print(f"Sample Mean           = {mean:.2f}")
    print(f"Sample Std Dev        = {std:.2f}")
    print(f"Confidence Level      = {confidence * 100:.0f}%")
    print(f"Z-critical value      = {z_critical:.3f}")
    print(f"Margin of Error       = {margin_of_error:.2f}")
    print(f"Confidence Interval   = ({lower:.2f}, {upper:.2f})")

    # Interpretation
    print(f"\nInterpretation: We are {confidence*100:.0f}% confident that the true population mean lies between {lower:.2f} and {upper:.2f}.")

# Example usage:
np.random.seed(42)
data = np.random.normal(loc=100, scale=15, size=50)  # Example dataset

confidence_interval(data, confidence=0.95)
```

5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu0=100, mu1=105, sigma=10, n=30, alpha=0.05):
    """
    Visualize Type I and Type II errors for a one-tailed hypothesis test (right-tailed).

    Parameters:
    mu0 : float - Population mean under H0
    mu1 : float - Population mean under H1
    sigma : float - Known population standard deviation
    n : int - Sample size
    alpha : float - Significance level
    """
    # Standard error
    se = sigma / np.sqrt(n)

    # Critical value (right-tailed test)
    z_critical = norm.ppf(1 - alpha)
    x_critical = mu0 + z_critical * se

    # Beta (Type II error): probability of accepting H0 when H1 is true
    z_beta = (x_critical - mu1) / se
    beta = norm.cdf(z_beta)

    # X-axis range
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)

    # Distributions
    y_h0 = norm.pdf(x, mu0, se)
    y_h1 = norm.pdf(x, mu1, se)

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(x, y_h0, label='Null Hypothesis (H₀)', color='blue')
    plt.plot(x, y_h1, label='Alternative Hypothesis (H₁)', color='green')

    # Shade Type I Error region (α)
    plt.fill_between(x, y_h0, where=(x >= x_critical), color='red', alpha=0.3, label='Type I Error (α)')

    # Shade Type II Error region (β)
    plt.fill_between(x, y_h1, where=(x <= x_critical), color='orange', alpha=0.3, label='Type II Error (β)')

    # Vertical line for critical value
    plt.axvline(x_critical, color='black', linestyle='--', label=f'Critical Value = {x_critical:.2f}')

    # Labels and legend
    plt.title('Visualization of Type I and Type II Errors')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Print summary
    print(f"Critical Value (Sample Mean Threshold): {x_critical:.2f}")
    print(f"Type I Error (α): {alpha:.2f}")
    print(f"Type II Error (β): {beta:.4f}")
    print(f"Power of the test (1 - β): {1 - beta:.4f}")
```

9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance?

```python
import numpy as np
from scipy.stats import t

def confidence_interval(sample, confidence=0.95):
    """
    Calculate the confidence interval for the sample mean.

    Parameters:
    - sample: list or numpy array of sample data
    - confidence: confidence level (default = 0.95 for 95%)

    Returns:
    - (mean, lower_bound, upper_bound)
    """
    sample = np.array(sample)
    n = len(sample)
    mean = np.mean(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)
    t_critical = t.ppf((1 + confidence) / 2, df=n - 1)

    margin_of_error = t_critical * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return mean, lower_bound, upper_bound

    # Example Usage
    data = [20, 22, 19, 21, 23, 20, 18, 24, 22]
    mean, ci_low, ci_high = confidence_interval(data, confidence=0.95)

    print(f"Sample Mean: {mean:.2f}")
    print(f"95% Confidence Interval: ({ci_low:.2f}, {ci_high:.2f})")
```

10. Write a Python program to calculate the margin of error for a given confidence level using sample data?

```python
import numpy as np
from scipy.stats import t

def calculate_margin_of_error(sample, confidence=0.95):
    """
    Calculate the margin of error for the sample mean.

    Parameters:
    - sample: list or array of numeric data
    - confidence: confidence level (default is 0.95 for 95%)

    Returns:
    - margin_of_error (float)
    """
    sample = np.array(sample)
    n = len(sample)
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)
    std_err = sample_std / np.sqrt(n)

    # t critical value for two-tailed test
    t_critical = t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_critical * std_err
    return sample_mean, margin_of_error

# Sample data
data = [45, 47, 50, 52, 46, 49, 51, 48, 47, 50]

# Calculate margin of error at 95% confidence level
mean, moe = calculate_margin_of_error(data, confidence=0.95)

print(f"Sample Mean: {mean:.2f}")
print(f"Margin of Error (95% confidence): ±{moe:.2f}")
print(f"Confidence Interval: ({mean - moe:.2f}, {mean + moe:.2f})")
```

11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process?

```python
def bayes_theorem(prior_H, prob_E_given_H, prob_E_given_not_H):
    """
    Compute posterior probability using Bayes' Theorem.

    Parameters:
    - prior_H: P(H) - prior probability of the hypothesis
    - prob_E_given_H: P(E|H) - likelihood
    - prob_E_given_not_H: P(E|~H) - false positive rate

    Returns:
    - posterior: P(H|E) - updated probability after evidence
    """
    prior_not_H = 1 - prior_H
    prob_E = prob_E_given_H * prior_H + prob_E_given_not_H * prior_not_H
    posterior = (prob_E_given_H * prior_H) / prob_E
    return posterior

# Given values
prior_disease = 1 / 1000             # P(H)
prob_positive_given_disease = 0.99   # P(E|H)
prob_positive_given_healthy = 0.05   # P(E|~H)

# Compute posterior probability
posterior_disease_given_positive = bayes_theorem(
    prior_disease,
    prob_positive_given_disease,
    prob_positive_given_healthy
)

# Display result
print(f"Probability of having the disease given a positive test: {posterior_disease_given_positive:.4f}")
```

12. Perform a Chi-square test for independence between two categorical variables in Python?

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# ---------------------------
# Step 1: Create a Contingency Table
# ---------------------------
# Example: Survey of gender vs. product preference
#           Product A  Product B
# Male         20         30
# Female       25         25

data = [[20, 30],
        [25, 25]]

# Optional: Use DataFrame for clarity
df = pd.DataFrame(data, columns=["Product A", "Product B"], index=["Male", "Female"])
print("Contingency Table:")
print(df)

# ---------------------------
# Step 2: Perform Chi-square Test
# ---------------------------
chi2, p, dof, expected = chi2_contingency(df)

# ---------------------------
# Step 3: Output the Results
# ---------------------------
print("\n--- Chi-square Test Results ---")
print(f"Chi-square Statistic: {chi2:.4f}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value: {p:.4f}")
print("\nExpected Frequencies:")
print(pd.DataFrame(expected, index=df.index, columns=df.columns))

# ---------------------------
# Step 4: Interpretation
# ---------------------------
alpha = 0.05
print("\n--- Conclusion ---")
if p < alpha:
    print("Reject the null hypothesis: There is a significant association between the variables.")
else:
    print("Fail to reject the null hypothesis: No significant association between the variables.")
```
